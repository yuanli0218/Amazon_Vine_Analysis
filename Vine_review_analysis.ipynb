```python
import pandas as pd
```

```python
# Read vine_data.csv file
Vine_Data_to_load = "Resources/vine_data_latest.csv"
vine_df = pd.read_csv(Vine_Data_to_load)
vine_df
```

```python
#Records with total_votes>20
vine_greaterthan20 = vine_df[(vine_df["total_votes"]>=20)]
vine_greaterthan20
```

```python
vine_greaterthan20["help_votes_percent"] =vine_greaterthan20["helpful_votes"]/vine_greaterthan20["total_votes"]
vine_greaterthan20
```

```python
# Rows where the helpful_votes/total_votes is greater than 50%
vine_votes = vine_greaterthan20[(vine_greaterthan20["help_votes_percent"]>=0.5)]
vine_votes
```

```python
# Total number of votes where helpful_votes/total_votes is greater than 50%
vine_votes["review_id"].count()
```

```python
# Number of paid reviews
vine_paid = vine_votes[(vine_votes["vine"]=='Y')]
vine_paid["review_id"].count()
```

```python
# Number of unpaid reviews
vine_unpaid = vine_votes[(vine_votes["vine"]=='N')]
vine_unpaid["review_id"].count()
```

```python
# Number of Vine paid 5 star reviews
vine_paid_5star = vine_paid[vine_paid["star_rating"] == 5]
vine_paid_5star["review_id"].count()
```

```python
# Vine 5 star paid percent
vine_5star_paid_percent = (vine_paid_5star["review_id"].count()/vine_paid["review_id"].count())*100
vine_5star_paid_percent
```

```python
# Number of Vine Unpaid 5 Star
vine_unpaid_5star = vine_unpaid[vine_unpaid["star_rating"] == 5]
vine_unpaid_5star["review_id"].count()
```

```python
# Vine Unpaid 5 star percent
vine_5star_unpaid_percent = (vine_unpaid_5star["review_id"].count()/vine_unpaid["review_id"].count())*100
vine_5star_unpaid_percent
```
